```python
import numpy as np 
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
# Melakukan Pembacaan Dataset 
dataset = pd.read_csv('Car_Evaluation.csv')

# Menampilkan dataset
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

Pre-Processing

```python
#Checking Null Values
dataset.isna().sum()
```

```python
#Checking Duplicates
data_duplicates_rows=dataset[dataset.duplicated()]
data_duplicates_rows
```

```python
#Removing Duplicates in any
dataset=dataset.drop_duplicates()
dataset
```

```python
en = LabelEncoder()

dataset['buying'] = en.fit_transform(dataset['buying'])
dataset['maint'] = en.fit_transform(dataset['maint'])
dataset.doors.replace(('5more'),('5'),inplace=True)
dataset.persons.replace(('more'),('5'),inplace=True)
dataset['lug_boot'] = en.fit_transform(dataset['lug_boot'])
dataset['safety'] = en.fit_transform(dataset['safety'])
dataset['class'] = en.fit_transform(dataset['class'])
```

```python
dataset
```

```python
#split dataset
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 6].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
#Feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# SVM using Linear kernel
svm_linear= SVC(kernel='linear',random_state=42)
```

```python
svm_linear.fit(X_train,y_train)
```

```python
y_pred_test_svm=svm_linear.predict(X_test)
```

```python
accuracy_score_svm=accuracy_score(y_test,y_pred_test_svm)
accuracy_score_svm
```

```python

print('\nHASIL KLASIFIKASI:\n\n', classification_report(y_test, y_pred_test_svm, zero_division=1))
```
